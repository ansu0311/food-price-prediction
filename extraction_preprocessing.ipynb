```python
import pandas as pd
df = pd.read_csv("./wfp_food_prices_ind.csv")
df
```

```python
df=df[1:][:]
df
```

```python
df = df.dropna()
df.reset_index(drop = True, inplace = True)
df.shape
```

```python
df=df.drop(labels = 'usdprice',axis=1)
df=df.drop(labels = 'currency',axis=1)
df=df.drop(labels = 'pricetype',axis=1)
df
```

```python
df=df.rename({"admin1":"province","admin2":"district"},axis = 1)
df
```

```python
import numpy as np

size=len(df['date'])
year_arr=np.zeros((size,1))
month_arr=np.zeros((size,1))
i=0
for date in df['date']:
    darr=date.split('-')
    year_arr[i] = darr[0]
    month_arr[i] = darr[1]
    i +=1

df['year']=year_arr.astype(int)
df['month']=month_arr.astype(int)
df=df.drop(labels='date',axis=1)

df
```

```python
dict_category ={'cereals and tubers':1, 'miscellaneous food':2, 'oil and fats':3,
       'pulses and nuts':4, 'vegetables and fruits':5, 'milk and dairy':6}
```

```python
size=len(df['category'])
catg=df['category'].unique()

catgarr=list(dict_category.keys())
carr=dict_category
for i in range (len(catg)):
  a=np.zeros((size,1))
  exec('catgarr[i] = a')


for j in range (size):
  cval=df['category'][j]
  ind=carr.get(cval)
  catgarr[ind-1][j]=1

r=0
for c in catg:
  df[c]=catgarr[r].astype(int)
  r+=1

df=df.drop(labels='category',axis=1) 
df
```
