```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression().fit(X_train, y_train)
print("Score",reg.score(X_test, y_test))
y_predict = reg.predict(X_test)
```

```python
import math
def accuracy(pred,test):
  cnt=0
  length=len(pred)
  for i in range (length):
    addv=math.ceil(pred[i]*1.25)
    subv=math.ceil(pred[i]*0.75)

    if(pred[i]==test[i]):
      cnt+=1
    elif(test[i]>=pred[i] and test[i]<=addv):
      cnt+=1
    elif(test[i]<=pred[i] and test[i]>=subv):
      cnt+=1
  
  acc=cnt/length
  return acc
```

```python
y_test_result=y_test.values
pred = np.ceil(y_predict)
test= np.ceil(y_test_result)
acc=accuracy(pred,test)
print(acc*100,"%")
```
